```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/anike/Downloads/DataScience_salaries_2024.csv")
df.head()
```

```python
# Create a 1x3 grid of plots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Violin Plot 1: Experience Level vs Salary
sns.violinplot(x='experience_level', y='salary_in_usd', data=df, ax=axes[0], inner='quart', palette='muted')
axes[0].set_title('Experience Level vs Salary')
axes[0].set_xlabel('Experience Level')
axes[0].set_ylabel('Salary in USD')

# Violin Plot 2: Employment Type vs Salary
sns.violinplot(x='employment_type', y='salary_in_usd', data=df, ax=axes[1], inner='quart', palette='muted')
axes[1].set_title('Employment Type vs Salary')
axes[1].set_xlabel('Employment Type')
axes[1].set_ylabel('Salary in USD')

# Violin Plot 3: Company Size vs Salary
sns.violinplot(x='company_size', y='salary_in_usd', data=df, ax=axes[2], inner='quart', palette='muted')
axes[2].set_title('Company Size vs Salary')
axes[2].set_xlabel('Company Size')
axes[2].set_ylabel('Salary in USD')

# Adjust layout for better spacing
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Create the violin plot for Job Title vs Salary in USD
plt.figure(figsize=(20, 12))
sns.violinplot(x='job_title', y='salary_in_usd', data=df, inner='quart', palette='muted')

# Set plot labels and title
plt.title('Job Title vs Salary')
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a 1x2 grid of plots (2 plots in a single row, since we removed the 3rd plot)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Company Location vs Salary (Plot 1)
sns.violinplot(x='company_location', y='salary_in_usd', data=df, ax=axes[0], inner='quart', palette='muted')
axes[0].set_title('Company Location vs Salary')
axes[0].set_xlabel('Company Location')
axes[0].set_ylabel('Salary in USD')

# Work Year vs Salary (Plot 2)
sns.violinplot(x='work_year', y='salary_in_usd', data=df, ax=axes[1], inner='quart', palette='muted')
axes[1].set_title('Work Year vs Salary')
axes[1].set_xlabel('Work Year')
axes[1].set_ylabel('Salary in USD')
# Rotate x-axis labels for better readability
axes[1].tick_params(axis='x', rotation=90)

# Adjust layout for better spacing between plots
plt.tight_layout()

# Display the plots
plt.show()
```
